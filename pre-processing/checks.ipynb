```python
import pandas as pd

# Make sure pandas does not truncate column display
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Read the first sheet
file_path = "output.xlsx" # Vintage radiator classification data.xlsx
df = pd.read_excel(file_path, sheet_name=0)

# Print all column names (first row headers)
# print(list(df.columns))
```

```python
for col in df.columns:
    print(col)
```

### Missing Values

```python
# Count missing values in each column
missing_per_column = df.isnull().sum()

print(missing_per_column)
```

```python
missing_columns = df.isnull().sum()
missing_columns = missing_columns[missing_columns > 0]

missing_df = missing_columns.reset_index()
missing_df.columns = ['Column Name', 'Missing Values']

print(missing_df)
```

### 'Tags' column

```python
# Drop missing values in the 'Tags' column
tags_series = df["Tags"].dropna()

# Create an empty set to store unique tag parts
unique_tags = set()

# Iterate through each cell in the Tags column
for item in tags_series:
    # Split by English comma
    parts = str(item).split(',')
    
    # Strip whitespace and add to set
    for part in parts:
        clean_part = part.strip()
        if clean_part:  # avoid empty strings
            unique_tags.add(clean_part)

# Print all unique parts
for tag in sorted(unique_tags):
    print(tag)
```

### 8 columns rep

```python
# The 8 columns to check for exact matches
columns_of_interest = [
    "Section Length (mm)",
    "Leg Section Depth (mm)",
    "Mid Section Depth (mm)",
    "Leg Section Height (mm)",
    "Mid Section Height (mm)",
    "Pipe Centre Top To Floor (mm)",
    "Pipe Centre Bottom To Floor (mm)",
    "Inter Axis Distance (mm)"
]

# Drop rows with any missing values in these 8 columns
df_subset = df[columns_of_interest].dropna()

# Find duplicates based only on these 8 columns
duplicates_mask = df_subset.duplicated(keep=False)

# Get the indices of duplicates in df_subset
duplicate_indices = df_subset.index[duplicates_mask]

# Use these indices to extract rows from the original df
duplicates = df.loc[duplicate_indices]

# Output only the first three columns of each duplicate row
first_three_columns = df.columns[:3]  # or specify explicitly: ["Column1", "Column2", "Column3"]
print(duplicates[first_three_columns])
```

```python
# Drop missing values in Tags column
cols = df[["Family", "Manufacturer", \
           "Leg Section Height (mm)", "Section Length (mm)", "Leg Section Depth (mm)", 'Nipple Size Top (")', \
            "Column", "Row", "Wall", "Window", "Ornate", "Plain", "Round top", "Flat top", "Flue"]]

# Find duplicated cols
duplicates = df[df.duplicated(subset=cols, keep=False)]

if not duplicates.empty:
    print("Duplicate rows found:")
    print(duplicates)
else:
    print("No duplicate rows found.")
```
