```python
from graphviz import Digraph

# Create canvas
dot = Digraph('Beeston_Radiator_Flowchart', format='png')
dot.attr(
    rankdir='TB',
    size='20',
    dpi='300',
    bgcolor='white',
    fontname='Arial',
    fontsize='14'
)

# Graph spacing
dot.graph_attr.update(
    nodesep='0.6',
    ranksep='0.3',
    splines='ortho'
)

# Node style
dot.node_attr.update(
    fontname='Arial',
    fontsize='12',
    penwidth='2'
)

# Edge style
dot.edge_attr.update(
    fontname='Arial',
    fontsize='10',
    penwidth='1.5',
    arrowsize='0.8'
)

# Helper function
def add_node(name, label, shape, color):
    dot.node(name, label, shape=shape, style='filled', fillcolor=color)

# START
add_node('START', 'START', 'oval', '#a2d2ff')

# Question nodes
questions = [
    ('Q1', 'Q1: Top Shape and Decoration?\n(Plain, Flat top, Scroll, Round top)'),
    ('Q2', 'Q2: Section Length (mm)?'),
    ('Q3', 'Q3: Leg Section Depth (mm)?'),
    ('Q4', 'Q4: Mid Section Height (mm)?'),
    ('Q5', 'Q5: Pipe Centre Bottom to Floor (mm)?')
]

for q, label in questions:
    add_node(q, label, 'diamond', '#ffe066')

# Connect questions sequentially
dot.edge('START', 'Q1')

for i in range(len(questions) - 1):
    dot.edge(questions[i][0], questions[i+1][0])

# Database check
add_node('DB1', 'Database Check\nMatching Models', 'box', '#bde0fe')
dot.edge('Q5', 'DB1')

# Identified
add_node('ID1', 'Radiator Identified', 'oval', '#90e0ef')
dot.edge('DB1', 'ID1', label='Single Match')

# Additional questions
add_node('Q6', 'Q6: Easy Clean\nDesign?', 'diamond', '#ffe066')
add_node('Q7', 'Q7: Nipple Top/\nBottom Proportion?', 'diamond', '#ffe066')
add_node('Q8', 'Final Distinction\nAmong Beeston Families', 'diamond', '#ffe066')

dot.edge('DB1', 'Q6', label='Multiple Matches')
dot.edge('Q6', 'Q7', label='Next')
dot.edge('Q7', 'Q8', label='Next')

# Final families
families = [
    ('F1', 'Duchess'),
    ('F2', 'Decorated'),
    ('F3', 'Princess\n(including Royal School)')
]

for f, label in families:
    add_node(f, label, 'oval', '#90e0ef')
    dot.edge('Q8', f)

# Render
dot.render('beeston_radiator_flowchart', view=True, cleanup=False)
```
